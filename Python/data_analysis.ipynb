```python
from datetime import date, timedelta

# Configuration
start_date = date(2024, 1, 1)
first_sprint_end = date(2024, 1, 13)
final_date = date(2027, 12, 31)

result = []
sprint_number = 67

current_start = start_date

# Handle first sprint (13 days)
current_end = first_sprint_end
result.append({
    'sprint': f'DR-Sprint-{sprint_number}',
    'start': current_start,
    'end': current_end
})
sprint_number += 1
current_start = current_end + timedelta(days=1)

# Handle subsequent sprints (14 days each)
while current_start <= final_date:
    current_end = current_start + timedelta(days=13)  # 14 days inclusive
    if current_end > final_date:
        current_end = final_date
    result.append({
        'sprint': f'DR-Sprint-{sprint_number}',
        'start': current_start,
        'end': current_end
    })
    sprint_number += 1
    current_start = current_end + timedelta(days=1)

# Print or process as needed
for entry in result:
    print(f"{entry['sprint']}: {entry['start']} to {entry['end']}")
```
